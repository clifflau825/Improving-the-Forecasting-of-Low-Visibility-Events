This script visualises the relationship between day of the year (1-366) and transformed day of year by applying sin(2*pi*(day_of_year-1)/366) and cos(2*pi*(day_of_year-1)/366). Figure 3.1 in the report is produced by this script.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
day_of_year = np.arange(1,367)
```

```python
day_of_year_sin = np.sin(2*np.pi*(day_of_year-1)/366)
day_of_year_cos = np.cos(2*np.pi*(day_of_year-1)/366)
```

```python
plt.figure(figsize=(8,4))
plt.plot(day_of_year,day_of_year_sin, label="day_of_year_sin")
plt.plot(day_of_year,day_of_year_cos, label="day_of_year_cos")
plt.legend()
plt.xlabel("day_of_year")
plt.ylabel("Transformed day_of_year")
plt.title("Relationship between transformed day_of_year and day_of_year")
plt.savefig("/Users/clifflau/Desktop/day_of_year", dpi=300)
```
