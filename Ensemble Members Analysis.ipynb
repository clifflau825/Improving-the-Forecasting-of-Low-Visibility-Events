This is a script for analysing the deviations in forecasts among different ensemble members in the EUPPBench forecasts and reforecasts.

```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "./"
export_path = "./"
```

```python
vis_reforecast = xr.open_dataarray(data_path + "vis_reforecasts.nc")
t_reforecast = xr.open_dataarray(data_path + "t_reforecasts.nc")
vis_forecast = xr.open_dataarray(data_path + "vis_forecasts.nc")
t_forecast = xr.open_dataarray(data_path + "t_forecasts.nc")
```

```python
vis_reforecast_array = np.array(vis_reforecast[:, :, 0:118, :]) # consider first 119 stations only
vis_reforecast_array[vis_reforecast_array>100000] = np.nan # remove outliers
t_reforecast_array = np.array(t_reforecast[:, :, 0:118, :]) # consider first 119 stations only
vis_forecast_array = np.array(vis_forecast[:, :, 0:118, :]) # consider first 119 stations only
vis_forecast_array[vis_forecast_array>100000] = np.nan # remove outliers
t_forecast_array = np.array(t_forecast[:, :, 0:118, ]) # consider first 119 stations only
```

```python
plt.figure(figsize=(7,5))
for i in range(0,11):
    plt.hist(vis_reforecast_array[:,i,:,:].flatten()/1000, bins=15, label="Member "+str(i+1))
plt.xlim(0,100)
plt.xlabel("Forecasted Visibility (km)")
plt.title("Distributions of NWP Visibitity Reforecasts by Members \n(1997-2018)")
plt.legend()
#plt.savefig(export_path + "distribution_vis_reforecast_member", dpi=300)
```

```python
plt.figure(figsize=(7,5))
for i in range(0,11):
    plt.hist(vis_forecast_array[:,i,:,:].flatten()/1000, bins=15, label="Member "+str(i+1))
plt.xlim(0,100)
plt.xlabel("Forecasted Visibility (km)")
plt.title("Distributions of NWP Visibitity Forecasts by Members \n(2017-2018)")
plt.legend()
#plt.savefig(export_path + "distribution_vis_forecast_member", dpi=300)
```

```python
plt.figure(figsize=(7,5))
for i in range(0,11):
    plt.hist(t_reforecast_array[:,i,:,:].flatten()/1000, bins=15, label="Member "+str(i+1))
plt.xlabel("Forecasted Air Temperature (K)")
plt.title("Distributions of NWP Air Temperature Reforecasts by Members \n(1997-2018)")
plt.legend()
#plt.savefig(export_path + "distribution_t_reforecast_member", dpi=300)
```

```python
plt.figure(figsize=(7,5))
for i in range(0,11):
    plt.hist(t_forecast_array[:,i,:,:].flatten()/1000, bins=15, label="Member "+str(i+1))
plt.xlabel("Forecasted Air Temperature (K)")
plt.title("Distributions of NWP Air Temperature Forecasts by Members \n(2017-2018)")
plt.legend()
#plt.savefig(export_path + "distribution_t_forecast_member", dpi=300)
```
