This is a script for analysing the standard deviations of NWP outputs among different ensemble members in the EUPPBench forecasts and reforecasts for the probability distributions of ensemble forecasts. This script produces the plots in Section 3.2.1 of the report.

```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Change for own use
ROOT = "ROOT/"
data_path = ROOT+"preprocessed_data/"
```

```python
df_train = pd.read_csv(data_path + "df_training.csv", dtype={'time_in_day':'category'})
df_test = pd.read_csv(data_path + "df_test.csv", dtype={'time_in_day':'category'})
```

```python
plt.figure(figsize=(8,6))
plt.hist(df_test['vis_ensemble_mean'], bins=20)
plt.xlim(0,100000)
plt.ylim(0,120000)
plt.xlabel("Visibility Forecasts (m)")
plt.ylabel("Frequency")
plt.title("Distribution of Ensemble Means of NWP Visibility Forecasts in 2018")
#plt.savefig("./distribution_nwp", dpi=300)
```

```python
plt.figure(figsize=(8,6))
plt.hist(df_test['vis_obs'], bins=20)
plt.xlim(0,100000)
plt.ylim(0,120000)
plt.xlabel("Visibility Observation (m)")
plt.ylabel("Frequency")
plt.title("Distribution Observaed Visibility in 2018")
#plt.savefig("./distribution_obs", dpi=300)
```

```python
plt.figure(figsize=(8,6))
plt.hist(np.log(df_test['vis_obs']+0.001), bins=20)
#plt.xlim(0,100000)
#plt.ylim(0,120000)
plt.xlabel("Visibility Observation (m)")
plt.ylabel("Frequency")
plt.title("Distribution Observaed Visibility in 2018")
#plt.savefig("./distribution_obs", dpi=300)
```

```python
fig, axs = plt.subplots(3,2, figsize=(13,17))
axs[0,0].hist(df_train['vis_ensemble_std']/1000, bins=30)
axs[0,0].set_xlim(0,30)
axs[0,0].set_title("Visibility Reforecasts")
axs[0,0].set_xlabel("Std of Members' Visibility Reforecasts (km)")
axs[0,0].set_ylabel("Frequency")
axs[0,1].hist(df_test['vis_ensemble_std']/1000, bins=30)
axs[0,1].set_xlim(0,30)
axs[0,1].set_title("Visibility Forecasts")
axs[0,1].set_xlabel("Std of Members' Visibility Forecasts (km)")
axs[0,1].set_ylabel("Frequency")
axs[1,0].hist(df_train['t_ensemble_std'], bins=30)
axs[1,0].set_xlim(0,7)
axs[1,0].set_title("Air Temperature Reforecasts")
axs[1,0].set_xlabel("Std of Members' Air Temperature Reforecasts (K)")
axs[1,0].set_ylabel("Frequency")
axs[1,1].hist(df_test['t_ensemble_std'], bins=30)
axs[1,1].set_xlim(0,7)
axs[1,1].set_title("Air Temperature Forecasts")
axs[1,1].set_xlabel("Std of Members' Air Temperature Forecasts (K)")
axs[1,1].set_ylabel("Frequency")
axs[2,0].hist(df_train['q_ensemble_std'], bins=30)
axs[2,0].set_xlim(0,3)
axs[2,0].set_title("Specific Humidity Reforecasts")
axs[2,0].set_xlabel("Std of Members' Specific Humidity Reforecasts (g/kg)")
axs[2,0].set_ylabel("Frequency")
axs[2,1].hist(df_test['q_ensemble_std'], bins=30)
axs[2,1].set_xlim(0,3)
axs[2,1].set_title("Specific Humidity Forecasts")
axs[2,1].set_xlabel("Std of Members' Specific Humidity Forecasts (g/kg)")
axs[2,1].set_ylabel("Frequency")
fig.suptitle("Distribution of Standard Deviations of Forecasts and Reforecasts from Ensemble Members", fontsize=15)
#plt.savefig("./distribution_of_std_1", dpi=300)
```

```python
fig, axs = plt.subplots(3,2, figsize=(13,17))
axs[0,0].hist(df_train['tcc_ensemble_std'], bins=30)
#axs[3,0].set_xlim(0,3)
axs[0,0].set_title("Total Cloud Cover Reforecasts")
axs[0,0].set_xlabel("Std of Members' Total Cloud Cover Reforecasts")
axs[0,0].set_ylabel("Frequency")
axs[0,1].hist(df_test['tcc_ensemble_std'], bins=30)
#axs[3,1].set_xlim(0,3)
axs[0,1].set_title("Total Cloud Cover Forecasts")
axs[0,1].set_xlabel("Std of Members' Total Cloud Cover Forecasts")
axs[0,1].set_ylabel("Frequency")
axs[1,0].hist(df_train['u10_ensemble_std'], bins=30)
axs[1,0].set_xlim(0,6)
axs[1,0].set_title("u10 Reforecasts")
axs[1,0].set_xlabel("Std of Members' u10 Reforecasts (m/s)")
axs[1,0].set_ylabel("Frequency")
axs[1,1].hist(df_test['u10_ensemble_std'], bins=30)
axs[1,1].set_xlim(0,6)
axs[1,1].set_title("u10 Forecasts")
axs[1,1].set_xlabel("Std of Members' u10 Forecasts (m/s)")
axs[1,1].set_ylabel("Frequency")
axs[2,0].hist(df_train['v10_ensemble_std'], bins=30)
axs[2,0].set_xlim(0,6)
axs[2,0].set_title("v10 Reforecasts")
axs[2,0].set_xlabel("Std of Members' v10 Reforecasts (m/s)")
axs[2,0].set_ylabel("Frequency")
axs[2,1].hist(df_test['v10_ensemble_std'], bins=30)
axs[2,1].set_xlim(0,6)
axs[2,1].set_title("v10 Forecasts")
axs[2,1].set_xlabel("Std of Members' v10 Forecasts (m/s)")
axs[2,1].set_ylabel("Frequency")
fig.suptitle("Distribution of Standard Deviations of Forecasts and Reforecasts from Ensemble Members", fontsize=15)
#plt.savefig("./distribution_of_std_2", dpi=300)
```
